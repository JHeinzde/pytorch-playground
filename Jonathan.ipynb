```python
class PrintLogs(tf.keras.callbacks.Callback):
    def __init__(self, epochs):
        self.epochs = epochs
        self.pbar = tqdm(total=epochs, dynamic_ncols=True)

    def on_epoch_end(self, epoch, logs=None):
        if logs is not None:
            self.pbar.set_description("Loss: {:.4f}".format(logs.get('loss')))
        self.pbar.update(1)

    def on_train_end(self, logs=None):
        self.pbar.close()
```

# Autoencoder

```python
# Preprocessing
normal_df = df[df['label'] == le.transform(['benign'])[0]]

# Drop the 'label' column as we don't need it for the autoencoder
X = normal_df.drop(['label'], axis=1)

# Normalize the data
scaler = MinMaxScaler()
X = scaler.fit_transform(X)

# Define the dimensions of the autoencoder
input_dim = X.shape[1]  # number of columns in the dataframe
encoding_dim = 2  # dimension of the latent space

# Define the autoencoder layers
input_layer = Input(shape=(input_dim, ))
encoder_layer = Dense(encoding_dim, activation='relu')(input_layer)
decoder_layer = Dense(input_dim, activation='sigmoid')(encoder_layer)

# Create the autoencoder model
autoencoder = Model(inputs=input_layer, outputs=decoder_layer)

# Compile the autoencoder model
autoencoder.compile(optimizer='adam', loss='binary_crossentropy')

# Train the autoencoder model
epochs = 500
print_logs = PrintLogs(epochs=epochs)
autoencoder.fit(X, X, epochs=epochs, batch_size=8, shuffle=True, validation_split=0.1, verbose=0, callbacks=[print_logs])

# Create a separate encoder model to map data to the latent space
encoder = Model(inputs=input_layer, outputs=encoder_layer)

# Use the encoder to map the data to the 2D latent space
latent_representation = encoder.predict(scaler.transform(df.drop(['label'], axis=1)), verbose=0)
```

```python
# Create a DataFrame for the latent representation
latent_df = pd.DataFrame(latent_representation, columns=['Dimension 1', 'Dimension 2'])

# Add the labels to the DataFrame
latent_df['Label'] = le.inverse_transform(df['label'])

# Create the scatter plot
scatter_plot = alt.Chart(latent_df).mark_circle(size=60).encode(
    x='Dimension 1',
    y='Dimension 2',
    color='Label',
    tooltip=['Dimension 1', 'Dimension 2', 'Label']
).properties(
    width=800,
    height=800
).interactive()

scatter_plot
```

# AnoGAN

```python
X = df.drop('label', axis=1).values # remove the label column

# Normalize the data
scaler = StandardScaler()
X = scaler.fit_transform(X)

# Define the generator and discriminator networks
class Generator(nn.Module):
    def __init__(self):
        super(Generator, self).__init__()
        self.main = nn.Sequential(
            nn.Linear(2, 128),
            nn.ReLU(),
            nn.Linear(128, X.shape[1]),
            nn.Tanh(),
        )
    def forward(self, input):
        return self.main(input)

class Discriminator(nn.Module):
    def __init__(self):
        super(Discriminator, self).__init__()
        self.main = nn.Sequential(
            nn.Linear(X.shape[1], 128),
            nn.ReLU(),
            nn.Linear(128, 1),
            nn.Sigmoid(),
        )
    def forward(self, input):
        return self.main(input)

# Initialize the networks
generator = Generator()
discriminator = Discriminator()

# Define the loss function and optimizers
criterion = nn.BCELoss()
optimizerD = torch.optim.Adam(discriminator.parameters(), lr=0.00003)
optimizerG = torch.optim.Adam(generator.parameters(), lr=0.00003)

# Train the GAN
epochs = 100
pbar = tqdm(range(epochs), desc="Training progress")
for epoch in pbar:
    for i in range(X.shape[0]):
        # Train the discriminator
        discriminator.zero_grad()
        real_data = Variable(torch.Tensor(X[i]))
        output = discriminator(real_data)
        errD_real = criterion(output, Variable(torch.ones(1)))
        errD_real.backward()
        D_x = output.mean().item()

        noise = Variable(torch.randn(2))
        fake = generator(noise)
        output = discriminator(fake.detach())
        errD_fake = criterion(output, Variable(torch.zeros(1)))
        errD_fake.backward()
        D_G_z1 = output.mean().item()

        errD = errD_real + errD_fake
        optimizerD.step()

        # Train the generator
        generator.zero_grad()
        output = discriminator(fake)
        errG = criterion(output, Variable(torch.ones(1)))
        errG.backward()
        D_G_z2 = output.mean().item()
        optimizerG.step()

    pbar.set_postfix({"Discriminator Loss": errD.item(), "Generator Loss": errG.item()})
```

# VAE

```python
# Preprocessing
normal_df = df[df['label'] == le.transform(['benign'])[0]]

# Split the data into features and target
X = normal_df.drop('label', axis=1)

# Data normalization
scaler = StandardScaler()
X = scaler.fit_transform(X)

# Define the dimensions of the layers
input_dim = X.shape[1]
intermediate_dim = 64
latent_dim = 2

# Define the encoder
inputs = Input(shape=(input_dim,))
h = Dense(intermediate_dim, activation='relu')(inputs)
z_mean = Dense(latent_dim)(h)
z_log_sigma = Dense(latent_dim)(h)

def sampling(args):
    z_mean, z_log_sigma = args
    epsilon = K.random_normal(shape=(K.shape(z_mean)[0], latent_dim),
                              mean=0., stddev=0.1)
    return z_mean + K.exp(z_log_sigma) * epsilon

z = Lambda(sampling)([z_mean, z_log_sigma])

# Define the decoder
decoder_h = Dense(intermediate_dim, activation='relu')
decoder_mean = Dense(input_dim, activation='sigmoid')
h_decoded = decoder_h(z)
x_decoded_mean = decoder_mean(h_decoded)

# Define the VAE model
vae = Model(inputs, x_decoded_mean)

# Define the VAE loss
reconstruction_loss = losses.mean_squared_error(inputs, x_decoded_mean)
reconstruction_loss *= input_dim
kl_loss = 1 + z_log_sigma - K.square(z_mean) - K.exp(z_log_sigma)
kl_loss = K.sum(kl_loss, axis=-1)
kl_loss *= -0.6
vae_loss = K.mean(reconstruction_loss + kl_loss)
vae.add_loss(vae_loss)

# Compile and train the model
epochs = 100
print_logs = PrintLogs(epochs=epochs)
vae.compile(optimizer='adam')
vae.fit(X, X, epochs=epochs, batch_size=8, verbose=0, callbacks=[print_logs])

# Define the encoder model
encoder = Model(inputs, z_mean)

# Use the encoder to map the data to the 2D latent space
latent_representation = encoder.predict(scaler.transform(df.drop(['label'], axis=1)), verbose=0)
```

# Siamese Network

```python
import torch
from torch import nn
import torch.optim as optim

def get_triplets(df_tensor, labels, model, use_hard_triplets=False):
    unique_labels = labels.unique()

    # Select random anchor
    anchor_idx = np.random.randint(0, len(df_tensor))
    anchor_label = labels[anchor_idx]
    anchor = df_tensor[anchor_idx]

    if use_hard_triplets:
        # Compute embeddings for all samples
        embeddings = model.fc(df_tensor)
        
        # Calculate pairwise distances between all embeddings
        pairwise_dist = torch.cdist(embeddings, embeddings, p=2)

        # For each anchor, find the hard positive
        mask_anchor_positive = labels == anchor_label
        mask_anchor_positive[anchor_idx] = False  # Ensure positive example is not the same as anchor
        hard_positive_idx = pairwise_dist[anchor_idx][mask_anchor_positive].argmax().item()
        positive = df_tensor[hard_positive_idx]

        # For each anchor, find the hard negative
        mask_anchor_negative = labels != anchor_label
        hard_negative_idx = pairwise_dist[anchor_idx][mask_anchor_negative].argmin().item()
        negative = df_tensor[hard_negative_idx]
    else:
        # Select random positive example
        positive_idx = np.random.choice(labels[labels == anchor_label].index)
        while positive_idx == anchor_idx:  # Ensure positive example is not the same as anchor
            positive_idx = np.random.choice(labels[labels == anchor_label].index)
        positive = df_tensor[positive_idx]

        # Select negative example (maximize distance to normal)
        # if anchor_label != le.transform(['benign'])[0]:
        #     negative_idx = np.random.choice(labels[labels == le.transform(['benign'])[0]].index)
        # else:
        #     negative_idx = np.random.choice(labels[labels != le.transform(['benign'])[0]].index)
        # negative = df_tensor[negative_idx]

        # Select negative example (maximize class separation)
        negative_idx = np.random.choice(labels[labels != anchor_label].index)
        negative = df_tensor[negative_idx]

    return anchor, positive, negative


# Define the network architecture
class SiameseNetwork(nn.Module):
    def __init__(self):
        super(SiameseNetwork, self).__init__()
        self.fc = nn.Sequential(
            nn.Linear(100, 128),
            nn.ReLU(),
            nn.Linear(128, 64),
            nn.ReLU(),
            nn.Linear(64, 32),
            nn.ReLU(),
            nn.Linear(32, 16),
            nn.ReLU(),
            nn.Linear(16, 2)
        )

    def forward(self, input1, input2, input3):
        output1 = self.fc(input1)
        output2 = self.fc(input2)
        output3 = self.fc(input3)
        return output1, output2, output3

# Define the triplet loss
class TripletLoss(nn.Module):
    def __init__(self, margin=1.0):
        super(TripletLoss, self).__init__()
        self.margin = margin

    def forward(self, anchor, positive, negative):
        distance_positive = (anchor - positive).pow(2).sum() # .pow(.5)
        distance_negative = (anchor - negative).pow(2).sum() # .pow(.5)
        losses = torch.relu(distance_positive - 1.5*distance_negative + self.margin)
        return losses.mean()

# Initialize the network and the loss function
model = SiameseNetwork()
criterion = TripletLoss()
optimizer = optim.Adam(model.parameters())

# Data normalization
X = df.iloc[:, :-1].values
scaler = StandardScaler()
X = scaler.fit_transform(X)

df_tensor = torch.tensor(X, dtype=torch.float32)

epochs = 50
pbar = tqdm(range(epochs), desc="Training progress")
for epoch in pbar:
    running_loss = 0.0
    for i in range(len(df)):
        anchor, positive, negative = get_triplets(df_tensor, df['label'], model, use_hard_triplets=False)

        # zero the parameter gradients
        optimizer.zero_grad()

        # forward + backward + optimize
        output1, output2, output3 = model(anchor, positive, negative)
        loss = criterion(output1, output2, output3)
        loss.backward()
        optimizer.step()

        running_loss += loss.item()
    pbar.set_postfix({"loss": running_loss / len(df)})
```
